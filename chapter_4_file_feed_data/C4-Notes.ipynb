# file vs web data

filebased data 
* xls, csv dbf, spss
* static, downloaded files

webbased data
* xl, jspm, rss
* api, dynamic

no perfect distinction, choose what u have available.
recent? web data
trend? file data

* feed: list or dictionary

unstructured
* different data types, pictures etc.

but: 
* structure: is dependent on the source. bias? poor design?
