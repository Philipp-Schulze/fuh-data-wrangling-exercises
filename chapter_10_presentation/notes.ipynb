# Visu Forms

* absolute values vs relationships important
* line, bar, pie, scatter etc.

## Pie Charts

* proportion or shares
* eg: voting results
* start at 12m, clock wise
  
rules:
* SUM OF A WHOLE
* less than 5 categories
* highlight proportion good visible eg 3/4, 1/2

## Bar and Column Charts

* highlight proportion and relationships
* data not a whole, positive and negative
* point in time or data over time
* vertical or horizontal
* MUST START AT ZERO, otherwise visual lie

rules:
* bars begin at 0
* vertical: dense data
* horizontal: more variation, bigger labels

## Lines Charts

* RATE OF CHANGE eg over time
* dont need to start at zero
* dont scale y acis too big, the no differences readable
* choose x axis on the important time frame

Rules:
* data as values over time
* data should occupy lower 2/3 vertical
* max 4 lines, diff colors

## Scatter Charts

* TREND OR DEVIATION FROM IT
* sometimes trendline as benchmark

Rules:
* high volume -> trends, outliers
* benchmarks are highlighted
* most of data is background, few highlights
* opacity for better distinguishability between data points

## Maps

* geographic data
* not often used
* geopandas library

# Eloquent visus

key points:
* finicky details matter, costs vs effect see frontend stuff
* colors: no science, trust your eyes
* selecting scales: whole numbers, no sci notation, labels and units
* colors: one color per data category, matching, no more than 4 distinct ones
* annotate: label in graphics, highlight the relevant, add annotations
